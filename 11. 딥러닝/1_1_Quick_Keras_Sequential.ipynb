## Quick Linear Regression!

```
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(0,20))
y = x * 2 -1 

print(x)
print(y)
```

```
x.shape, y.shape
```

```
# 1번 청소 : 이미 만들어진 모델이 있다면 그 모델을 없애줘
keras.backend.clear_session()

# 2번 모델 선언
model = keras.models.Sequential()

# 3번 모델 블록 조립
model.add( keras.layers.Input(shape=(1,)) )
model.add( keras.layers.Dense(1) )

## 오리지널 Sequential API
# model.add( keras.layers.Dense(1, input_shape=(1,)) )

# 4번 컴파일 
model.compile(loss='mse',
              optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
print(y)
```

```
model.predict(x).reshape(-1)
```

```
## 모델링 : Sequential API

# 1번. 세션 클리어 : 기존의 모델이 남아있다면 지워줘!
keras.backend.clear_session()

# 2번. 레이어 블록을 쌓을 발판 생성 : 레고처럼 차곡차곡
model = keras.models.Sequential()

# 3번. 레이어 블록 조립
model.add( keras.layers.Input(shape=(1,)))
model.add( keras.layers.Dense(1))

# 4번. 컴파일
model.compile(loss='mse', optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
y_pred = model.predict(x)

y_pred[:5].reshape(-1)
```

```
y[:5]
```

## Quick Logistic Regression!

```
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(0,20)) 
y = np.array([0]*10 + [1]*10)
print(x)
print(y)
```

```
x.shape, y.shape
```

```
# 혹시 이미 그려둔 그래프가 있다면 날려줘!
keras.backend.clear_session()

# model에 순차적으로 레이어를 쌓아가겠다는 의도!
model = keras.models.Sequential()

# model에 인풋 값을 받는 레이어를 넣음
model.add( keras.layers.Input(shape=(1,)) )
# model에 Dense 레이어를 넣을거야 (최초의 레이어) : weight를 곱하고, bias를 더해주는 과정
model.add( keras.layers.Dense(1, activation='sigmoid') )


# 오리지널 Sequential API
# model.add( keras.layers.Dense(1, input_shape=(1,), activation='sigmoid') )

# 컴파일 해주렴
model.compile(loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics=['accuracy'])
```

```
# 데이터를 넣어서 학습시키자!
model.fit(x[:15], y[:15], epochs=10, verbose=1)

# 결과 출력해줘!
print(y[15:])
print(model.predict(x[15:]))
```

```
model.predict(x).reshape(-1)
```

```
y.reshape(-1)
```

```
## 모델링 : Sequential API

# 1번 세션 클리어
keras.backend.clear_session()

# 2번 레이어 블록 발판 생성 : 차곡차곡 쌓을거야
model = keras.models.Sequential()

# 3번 .add()
model.add( keras.layers.Input(shape=(1,)))
model.add( keras.layers.Dense(1, activation='sigmoid'))

# 4번 .compile()
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
model.predict(x)[:5].reshape(-1)
```

```
y[:5]
```
