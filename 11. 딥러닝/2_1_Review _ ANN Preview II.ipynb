```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
x = cancer.data
y = cancer.target
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
### train set / test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2022)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
y_train[:5]
```

```python
#####################
# 라이브러리 불러오기
#####################
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```python
########
# 모델링
########
keras.backend.clear_session()

il = keras.layers.Input(shape=(30,), name='input')
ol = keras.layers.Dense(1, activation='sigmoid', name='output')(il)

model = keras.models.Model(inputs = il, outputs = ol)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary() # 31 + 1(bias)
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs=10, verbose=1)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_test)
y_pred[:5]
```

# 히든 레이어 추가하여 모델링!

    히든 레이어 2개를 추가하여 다시 모델링!
    히든 레이어 노드 수 : 32
    히든 레이어 activation = 'relu'

```python
########
# 모델링
########
keras.backend.clear_session()

il = keras.layers.Input(shape=(30,), name='input')
hl1 = keras.layers.Dense(32, activation='relu', name='hidden1')(il)
hl2 = keras.layers.Dense(32, activation='relu', name='hidden2')(hl1)
ol = keras.layers.Dense(1, activation='sigmoid', name='output')(hl2)

model = keras.models.Model(inputs = il, outputs = ol)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs=10, verbose=1)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_test)
y_pred[:5]
```

```python
y_pred2 = np.where(y_pred >= 0.5, 1, 0)
y_pred2.reshape(-1)
```
