```python
from sklearn.datasets import load_wine
```

```python
data = load_wine()
```

```python
print(data.DESCR)
```

```python
x = data.data
y = data.target
```

```python
import pandas as pd
```

```python
pd.DataFrame(x, columns=data.feature_names).head()
```

## 문제

    - 1) 히든 레이어 없이 성능 보기
    - 2) 히든 레이어 추가하고 성능 보기 (자율)
3. 학습 : epochs=100, validation set 10%, 5번까지만 참을래!

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=2022)
```

```python
pd.DataFrame(x_train, columns=data.feature_names).describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
pd.DataFrame(x_train, columns=data.feature_names).describe()
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
n = len(set(y_train))
# n = len(np.unique(y_train))
n
```

```python
y_train = to_categorical(y_train, n)
y_test = to_categorical(y_test, n)
```

```python
pd.DataFrame(y_train, columns=data.target_names).head()
```

```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```python
x_train.shape, y_train.shape
```

```python
keras.backend.clear_session()

il = keras.layers.Input(shape=(13,))
ol = keras.layers.Dense(3, activation='softmax')(il)

model = keras.models.Model(il, ol)

model.compile(optimizer=keras.optimizers.Adam(),
              loss=keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])

model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=1, restore_best_weights=True)
```

```python
model.fit(x_train, y_train, epochs=100, verbose=1, callbacks=[es], validation_split=0.1)
```

```python
keras.backend.clear_session()

il = keras.layers.Input(shape=(13,))
hl = keras.layers.Dense(64, activation='relu')(il)
hl = keras.layers.Dense(64, activation='relu')(hl)
hl = keras.layers.Dense(64, activation='relu')(hl)
ol = keras.layers.Dense(3, activation='softmax')(hl)

model = keras.models.Model(il, ol)

model.compile(optimizer=keras.optimizers.Adam(),
              loss=keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])

model.summary()
```

```python
model.fit(x_train, y_train, epochs=100, verbose=1, callbacks=[es], validation_split=0.1)
```

```python
y_pred = model.predict(x_test)
```
